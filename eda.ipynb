```python
train_df = load_labels_df("train")
valid_df = load_labels_df("valid")
test_df = load_labels_df("test")
```

```python
def analyze_df(df):
    # display(df.head())
    display(df.sample(n=5, random_state=42))

    print("\n-- shape --\n")
    print(df.shape)

    print("\n-- data type --\n")
    print(df.dtypes)

    print("\n-- unique values count --\n")
    print(df.nunique())

    print("\n-- NaN values count --\n")
    print(df.isna().sum())

    # print("\n-- description --\n")
    # print(df.describe())
```

```python
analyze_df(train_df)
```

```python
analyze_df(valid_df)
```

```python
analyze_df(test_df)
```

# Box Distribution

```python
imageID_objCount = train_df['image_id'].value_counts()
objCount_imageCount = imageID_objCount.value_counts().sort_index(ascending=False)
plt.figure(figsize=(8, 4))
plt.bar(objCount_imageCount.index, objCount_imageCount.values)
plt.xlabel('Box Count Per Image')
plt.ylabel('Number of Images')
plt.title('Box Count Per Image Distribution')
plt.grid(axis='y')
plt.show()
```

```python
classID_objCount = train_df['class_id'].value_counts().sort_values(ascending=False)
print(classID_objCount.index)
colors = [classes[class_id]["rgba"] for class_id in classID_objCount.index]
names = [classes[class_id]["name"] for class_id in classID_objCount.index]
plt.figure(figsize=(8, 4))
plt.bar(names, classID_objCount.values, color=colors)
plt.xlabel('Class')
plt.ylabel('Box Count')
plt.xticks(ticks=classID_objCount.index, rotation=45, ha='right')
plt.title('Box Count By Class')
plt.grid(axis='y')
plt.show()
```

```python
x_count = train_df['x_center'].value_counts().sort_index()
y_count = train_df['y_center'].value_counts().sort_index()
plt.figure(figsize=(8, 4))
plt.plot(x_count.index, x_count.values, label="x_center")
plt.plot(y_count.index, y_count.values, label="y_center")
plt.xlabel('Value')
plt.ylabel('Number of Images')
plt.title('Box Position Distribution')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
width_count = train_df['width'].value_counts().sort_index()
height_count = train_df['height'].value_counts().sort_index()
plt.figure(figsize=(8, 4))
plt.plot(width_count.index, width_count.values, label="width")
plt.plot(height_count.index, height_count.values, label="height")
plt.xlabel('Size')
plt.ylabel('Number of Images')
plt.title('Box Size Distribution')
plt.legend()
plt.grid(axis='y')
plt.show()
```
